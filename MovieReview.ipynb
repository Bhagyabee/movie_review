```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet

def get_simple_pos(tag):

  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN
```

```python
from nltk import pos_tag
w = "better"
pos_tag([w])
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python


def clean_review(words):
  output_words =[]

  for w in words:
    if w.lower() not in stops:
      pos = pos_tag([w])
      clean_word =lm.lemmatize(w,pos= get_simple_pos(pos[0][1]))
      output_words.append(clean_word.lower())
  return output_words
```

```python
documents = [(clean_review(document),category) for document,category in documents]
```

```python
documents[0]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words =[]
for doc in training_documents:
  all_words += doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)

features = [i[0] for i in common]
features
```

```python
def get_feature_dict(words):
  current_features ={}
  words_set = set(words)
  for w in features:
    current_features[w]= w in words_set
  return current_features

```

```python
training_data = [ (get_feature_dict(doc),category)for doc , category in training_documents]
```

```python
testing_data = [ (get_feature_dict(doc),category)for doc , category in training_documents]
```

```python
from nltk import NaiveBayesClassifier
```

```python
clf = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(clf,testing_data)
```

```python
clf.show_most_informative_features(19)
```
